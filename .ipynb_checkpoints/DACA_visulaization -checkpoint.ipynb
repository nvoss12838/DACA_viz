So what does this show? The total hieght shows the number of people in the program at any time and the width of
and the width of the color shows the number of people that fall into each catagory. Could be more interesting as a
interactive plotly chart? Basicially copied from their stacker area example. 

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly
plotly.tools.set_credentials_file(username='nvoss12838', api_key='jMnraLGYmm3bcxziwCSp')
# Add original data
x=dfRenewals.index.to_timestamp()
y0_org=dfRenewals['No Renewal Pending']
y1_org=dfRenewals['Renewal Pending']
y2_org=dfRenewals['Renewal Granted']

# Add data to create cumulative stacked values
y0_stck=y0_org
# Add data to create cumulative stacked values

y1_stck=[y0+y1 for y0, y1 in zip(y0_org, y1_org)]
y2_stck=[y0+y1+y2 for y0, y1, y2 in zip(y0_org, y1_org, y2_org)]

# Make original values strings and add % for hover text
y0_txt=[str(ceil((float(y0)/(y0+y1_org[i]+y2_org[i]))*100))+'%' for i,y0 in enumerate(y0_org)]
y1_txt=[str(ceil((float(y1_org[i])/(y0+y1_org[i]+y2_org[i]))*100))+'%' for i,y0 in enumerate(y0_org)]
y2_txt=[str(ceil((float(y2_org[i])/(y0+y1_org[i]+y2_org[i]))*100))+'%' for i,y0 in enumerate(y0_org)]

No_Renewal = go.Scatter(
    x=x,
    y=y0_stck,
    text=y0_txt,
    hoverinfo='x+text',
    mode='lines',
    line=dict(width=0.5,
              color='red'),
    fill='tonexty',
    name='No Renewal'
)
Renewal_Pending = go.Scatter(
    x=x,
    y=y1_stck,
    text=y1_txt,
    hoverinfo='x+text',
    mode='lines',
    line=dict(width=0.5,
              color='yellow'),
    fill='tonexty',
    name='Renewal_Pending'
)
Renewal_Granted = go.Scatter(
    x=x,
    y=y2_stck,
    text=y2_txt,
    hoverinfo='x+text',
    mode='lines',
    line=dict(width=0.5,
              color='green'),
    fill='tonexty',
    name='Renewal Granted'
)
data = [No_Renewal,Renewal_Granted,Renewal_Pending]
layout = go.Layout(
    title='DACA Visualization',
    xaxis=dict(
        title='Time',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Number of People',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)
fig = go.Figure(data=data,layout=layout)

py.iplot(fig, filename='stacked-area-plot-hover')
```
